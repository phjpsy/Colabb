### ndarray 생성 – arange
- 파이썬 내장함수인 range는 list반환
- numpy arange는 ndarray 반환

```python
np.arange(10)  # list의 range와 같은 역할, integer로 0부터 9까지 배열추출
```

```python
np.arange(0, 1, 0.2) # float, step
```

```python
np.arange(20).reshape(4,5)
```

### ndarray 생성 – zeros, ones, empty
#### zeros : 0으로 가득찬 ndarray 생성|

```python
np.zeros(shape = (5)) #  요소 = 0인 vector 생성
```

```python
np.zeros((5,2)) # 5 by 2 zero matrix 생성
```

#### zeros : 1로 가득찬 ndarray 생성|

```python
np.ones(shape = (5))  #  요소 = 1인 vector 생성
```

```python
np.ones((4,3)) # 4 by 3 요소 = 1인 matrix 생성
```

#### empty : 메모리를 할당하여 새로운 배열 생성하나 memory를 초기화 하지 않음.

```python
np.empty(shape = (10,), dtype = np.int8)  #  요소 = 1인 vector 생성, int8
```

```python
np.empty((4,4))
```

### something_like
- 기존 ndarray의 shape 크기만큼 1 or 0 or empty array 반환

```python
t_matrix = np.arange(6).reshape(2,3)
np.ones_like(t_matrix)
```

```python
t_matrix1 = np.arange(8).reshape(2,4)
np.zeros_like(t_matrix1)
```

```python
t_matrix2 = np.arange(6).reshape(2,3)
np.empty_like(t_matrix2)
```

### identity
- 단위 행렬 생성

```python
np.identity(n = 3, dtype = np.int8)
```

```python
np.identity(n = 4)
```

### eye
- N X M 크기의 대각선이 1인 행렬 생성, K값에 따라 1의 시작점 변경


```python
np.eye(N = 2, M = 2, dtype = np.int8)
```

```python
np.eye(3)  # identity행렬과 같게 출력 
```

```python
np.eye(3, 6, k = 3)  # k는 시작점
```

### ndarray의 자료형
- 자료형, dtype은 ndarray가 데이터를 메모리에서 해석하기 위해 필  
   요한 정보를 담고있는 특수한 객체이다.
- ndarray의 각각의 요소가 가지는 data type을 통해 memory 크기
   가 결정된다 

```python
t_matrix = np.array([[4, 2.6, 7.2], [1, 3.1, 6]], dtype = np.float32 )
t_matrix
```

- nbyte : ndarray object의 메모리 크기 리턴

```python
t_matrix.nbytes
```

```python
np.array( [[1, 2.6, 3.2], [4, "5", 6]], dtype = np.int8 ).nbytes
```
